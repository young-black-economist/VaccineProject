```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv('D:/Dissertation/Vaccine/measles.csv')
df = df_raw[['Measles', 'Year']]
df.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)
ax.vlines(x=df.index, ymin=0, ymax=df.Measles, color='navy', alpha=0.7, linewidth=2)
ax.scatter(x=df.index, y=df.Measles, s=20, color='navy')

ax.set_title('Yearly Measles Cases in the U.S. 2002-2024', fontdict={'family':'serif','size':24})
ax.set_ylabel('Number of Measles Cases', fontdict={'family':'serif','size':18})
ax.set_xticks(df.index)
ax.set_xticklabels(df.Year, rotation=60, fontdict={'horizontalalignment': 'center', 'size':14})
ax.set_ylim(0, 1400)
for row in df.itertuples():
    ax.text(row.Index, row.Measles, s=round(row.Measles, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=16)

plt.savefig('D:/Dissertation/Vaccine/lolli.png')
```
